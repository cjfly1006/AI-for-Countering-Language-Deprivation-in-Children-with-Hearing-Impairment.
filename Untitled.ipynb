My first step was to teaching myself finger spelling to get some insight into who signing works and what type of data the algorithm would need to interoperate finger spelling. 
i then made a choice on the type of data i felt would be more effective in capturing the characters needed to fingerspell. 
The two main methods of gathering data were taking images of me signing each character or taking landmark data from videoes. I ended up using holistic data.


```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
import time
import mediapipe as mp
```

```python
mp_holistic = mp.solutions.holistic # Holistic model
mp_drawing = mp.solutions.drawing_utils # Drawing utilities
```

First I neaded to change the colour from the cv2's default BGR to RGB for the holistic model to make detections on the image, i then swaped this bak to the default once the model has made its detection.

```python
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Color conversion
    image.flags.writeable = False

    # Make prediction
    results = model.process(image)
    
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # Color conversion
    return image, results
    
```

This function helps present the necesary landmarks to the screen and capture the cordinets of the necesary landmarks for finger spelling.

```python
def draw_landmarks(image, results):
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
```

Then i writ some code to access my webcam and apply the holistics model imported from media pipe to each frame. this allowed me to test wether the characters could be captured using the holistic model, and practice finger spelling for when I wanted to recored data.

```python
cap = cv2.VideoCapture(0)
# Set mediapipe model
with mp_holistic.Holistic(min_detection_confidence = 0.5, min_tracking_confidence = 0.5) as holistic:
    while cap.isOpened():
        #Read feed
        ret, frame = cap.read()

        # Make detection
        image, results = mediapipe_detection(frame, holistic)

        # Draw landmarks
        draw_landmarks(image, results)
    
        #Show to screen
        cv2.imshow('OpenCV', image)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
    
```

This code allows me to access the land marks from the frames of my captured footage and will allow me to use them to train my character recognition model.

```python
def keypoints(results):
    lh = np.array([[res.x, res.y, res.z, res.visibility] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z, res.visibility] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([lh, rh])
```

```python
# Path for extracting data
data_path = ('SL_data')

# Actions
actions = np.array(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r',
                    's', 't', 'u', 'v', 'w', 'x', 'y', 'z'])

#videoes per action
no_sequence = 100

# frames in lenghth
sequence_len = 30


```

```python
for action in actions:
    for sequence in range(no_sequence):
        try:
            os.makedirs(os.path.join(data_path, action, str(sequence)))
        except:
            pass

```

This was the code for collecting data.

```python
cap = cv2.VideoCapture(0)
# Set mediapipe model
with mp_holistic.Holistic(min_detection_confidence = 0.5, min_tracking_confidence = 0.5) as holistic:

    #loop through actions
    for action in actions:
        for sequence in range(no_sequence):
            for frame_num in range(sequence_len):
                
                #Read feed
                ret, frame = cap.read()

                # Make detection
                image, results = mediapipe_detection(frame, holistic)
                print(results)

                # Draw landmarks
                draw_landmarks(image, results)

                if frame_num == 0:
                    
                    cv2.putText(image, 'start collecting', (120,200),
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 1, cv2.LINE_AA)
                    
                    cv2.putText(image, 'collecting frames for {} video number {}'.format(action, sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 2, cv2.LINE_AA)
                                        
                    cv2.waitKey(2000)
                    
                else:
                    cv2.putText(image, 'collecting frames for {} video number {}'.format(action, sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 2, cv2.LINE_AA)
                    cv2.imshow('openCV feed', image)

                #exports keypoints
                keypoint = keypoints(results)
                npy_path = os.path.join(data_path, action, str(sequence), str(frame_num))
                np.save(npy_path, keypoint)
    
                #Show to screen
                cv2.imshow('OpenCV', image)
        
                if cv2.waitKey(10) & 0xFF == ord('q'):
                     break
        cap.release()
        cv2.destroyAllWindows()
```

this was a neccesary step of pre-processing as the datatype of the data i had gathered was not applicable to a tensorflow model

```python
sequences, labels = [], []
for action in actions:
    for sequence in np.array(os.listdir(os.path.join(data_path, action))).astype(int):
        window = []
        for frame_num in range(sequence_len):
            res_path = os.path.join(data_path, action, str(sequence), f"{frame_num}.npy")
            if os.path.exists(res_path):
                res = np.load(res_path)
                # Ensure all frames have the same shape, e.g., reshape or pad if necessary
                if res.shape != (168):  # Define 'expected_shape' based on your data
                    res = np.resize(res, 168)  # Resize to expected_shape
                window.append(res)
            else:
                # Handle missing data
                window.append(np.zeros(168))  # Fill missing frames with zeros
        sequences.append(window)
        labels.append(label_map[action])

# Convert lists to arrays
sequences = np.array(sequences, dtype=np.float32)
labels = np.array(labels, dtype=np.int32)

# Convert arrays to TensorFlow variables
#X = tf.Variable(sequences)
```

i decided to use a long short term memory model as i wanted to capture sequences rather than still imaged and this way i could do both.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Masking
```

```python
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(30,168)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(actions.shape[0], activation='softmax'))
```

```python
# 1. New detection variables
sequence = []
sentence = []
predictions = []
threshold = 0.7

cap = cv2.VideoCapture(0)
# Set mediapipe model 
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():

        # Read feed
        ret, frame = cap.read()
        # Make detections
        image, results = mediapipe_detection(frame, holistic)
        print(results)        
        # Draw landmarks
        draw_landmarks(image, results)
        
        # 2. Prediction logic
        keypoint = keypoints(results)
        sequence.append(keypoint)
        sequence = sequence[-30:]
        
        if len(sequence) == 30:
            processed_sequence = []
            for frame_data in sequence:
                if frame_data.shape != (168):
                    frame_data = np.resize(frame_data, 168)
                processed_sequence.append(frame_data)
            processed_sequence = np.array(processed_sequence, dtype=np.float32)
            processed_sequence = np.expand_dims(processed_sequence, axis=0)
                        
            # Prediction
            res = model.predict(processed_sequence)[0]
            print(actions[np.argmax(res)])
            predictions.append(np.argmax(res))            
            
        #3. Vizual logic
            if np.unique(predictions[-10:])[0]==np.argmax(res): 
                if res[np.argmax(res)] > threshold: 
                    
                    if len(sentence) > 0: 
                        if actions[np.argmax(res)] != sentence[-1]:
                            sentence.append(actions[np.argmax(res)])
                    else:
                        sentence.append(actions[np.argmax(res)])

            if len(sentence) > 5: 
                sentence = sentence[-5:]

            # Viz probabilities
            image = prob_viz(res, actions, image, colors)
            
        cv2.rectangle(image, (0,0), (640, 40), (245, 117, 16), -1)
        cv2.putText(image, ' '.join(sentence), (3,30), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
        
        cv2.imshow('OpenCV Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
            
    cap.release()
    cv2.destroyAllWindows()
```
